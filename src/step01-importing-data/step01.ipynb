Read the data from a CSV file.

```julia
include("./utilities/read-files.jl")
dataframe = ReadDF("../../data/processed/DetrendedCov.csv")
first(dataframe, 5)
```

Plot the concentration data (N1 column) in red.

```julia
using Plots;
Plots.scatter(dataframe."Date", dataframe."N1", color="red", xlabel="X", ylabel="Y", 
              label=nothing, title="Observations - regular data and outliers")          
```

Plot the detrended concentration data (DetrendedN1 column) in blue.

```julia
using Plots;
Plots.scatter(dataframe."Date", dataframe."DetrendedN1", color="blue", xlabel="X", ylabel="Y", 
              label=nothing, title="Observations - regular data and outliers")
```
