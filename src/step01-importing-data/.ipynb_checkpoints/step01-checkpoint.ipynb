## Step01 - Importing Data

```julia
using Plots: scatter, scatter!
using CSV: read
using DataFrames: DataFrame
```

We load the data and converts used columns to float. Then the following cells display Base Plot and Detrended Plot.

```julia
# use CSV package to load the CSV file
dataframe = read("../../data/processed/DetrendedCov.csv", DataFrame)

# convert N1 column to floats
dataframe[!,:N1] = convert.(Base.Float64, dataframe[!,:N1])

# convert Date column to dates
dataframe[!,:Date] = convert.(Base.Float64, dataframe[!,:Date])
```

```julia
#Plot the Covid concentration data (N1 column) in blue.
scatter(dataframe."Date", dataframe."N1", color = "blue", xlabel = "Date", ylabel = "Concentration", 
label = "N1", title = "Observations")

#adds the detrended concentration data (DetrendedN1 column) in blue to the last plot.
scatter!(dataframe."Date", dataframe."DetrendedN1", color = "red", xlabel = "Date", ylabel = "Concentration", 
label = "N1 (Detrended)", title = "Observations")
```
